```python
def build_dataset(data,name):
    ''' build micro data 
    There is a gotcha here if we're not careful, which is that we are only using 
    microclimate data in this benchmark and ignoring the other available dataâ€”but
    the microclimate test data has some missing values!
    '''
    
    train = data['microclimat_train']
    train_y = data['labels']
    sub_format = data['submission_format']
    test = sub_format.join(data['microclimat_test'], how='left')
    test = test[microclimate_test.columns]
    
    return train, train_y, test 
    
    
```
